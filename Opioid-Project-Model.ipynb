```python
# split data
train_set,test_set = train_test_split(presc_rate_w_lobbying,test_size = 0.2,random_state = 42)
```

```python
train_set = train_set.drop_duplicates()
train_set = train_set[train_set['report_year'].notna()]
train_set = train_set[train_set.Tot_Clms > 0]
# train_set.drop('report_year',axis=1, inplace=True)
# train_set.drop('year',axis=1, inplace=True)
for each in to_remove:
    train_set.drop(each,axis=1, inplace=True)
```

```python
categorical_features = ['Year','Breakout','Prscrbr_State_Abrvtn']
```

```python
categorical_features
```

```python
category_values = {}

for feature in categorical_features:
    category_values[feature] = list(train_set[feature].value_counts().index)

lst1 = train_set['Breakout'].unique().tolist()
lst2 = train_set['Prscrbr_State_Abrvtn'].unique().tolist()
categorical_features = categorical_features + lst1 + lst2


print(category_values)
```

```python
def pre_process_data(df, categoical_values):
    new_df = df.copy(deep=True)
    for feature in df.columns:
        if feature in categorical_features:
            new_df[feature] = pd.factorize(df[feature], sort = True)[0]
    return new_df
    

```

```python
processed_train = pre_process_data(train_set, category_values)
processed_train.head()
```

## Conclusions

Numerically this model deos very well for the first two attempts. The difficuly with classifying this model as a succesful experiment lies in the way the data was pieced together. Initially tackling the data cleaning seems to be a simple task, but when combing data from different sources it can change the significance. Next steps to ensure that this model works as expected is really having the combined data be reviewed by a field expert. There is a correlation between the lobbying contributions an prescribing rate according to the model, but there are some problems with the way the data was merged. The lobbying data was only taken for a two year period and there was no "by-state" information found for this data, therefore each sample for prescribing rates was given the dollar amount for a year. For this reason the model runs the risk of being overfitted. Other data that was added was about the total amount of distributed opioids. In a future experiment, overall drug distribution should also be included. The data is skewed to have high correlation values. The third experiment shows a negative values, and my own interpretation of what that means is limitied still.

Attempts to correct some of these issues were done by running various feature selection. However, the problem we are attempting to fix is complex and there needds to be caution is how we deliver the results.
